```python
class Cancion:
    # Constructor: Aquí se definen los 3 atributos
    def __init__(self, titulo="", artista="", duracion=""):
        self.titulo = titulo
        self.artista = artista
        self.duracion = duracion

    # Función 1: Leer datos desde el teclado
    def leer_datos(self):
        print(f"\n--- Ingresando datos de la canción ---")
        self.titulo = input("Introduce el título: ")
        self.artista = input("Introduce el artista: ")
        self.duracion = input("Introduce la duración (ej. 3:45): ")

    # Función 2: Mostrar los datos en pantalla
    def mostrar_datos(self):
        print("\n--- Información de la Canción ---")
        print(f"Título:   {self.titulo}")
        print(f"Artista:  {self.artista}")
        print(f"Duración: {self.duracion}")
        print("---------------------------------")

    # Función 3: Un método extra para cumplir con los "3 funciones" solicitadas
    # Simula la acción de reproducir la canción
    def reproducir(self):
        print(f"Reproduciendo ahora: {self.titulo} de {self.artista}...")

# --- Creación de Objetos ---

# Objeto 1: Usando leer_datos()
cancion1 = Cancion()
print("Configurando Cancion1:")
cancion1.leer_datos()

# Objeto 2: Datos de ejemplo directos
cancion2 = Cancion("Bohemian Rhapsody", "Queen", "5:55")

# --- Ejecución de Funciones y Mensajes ---

# Mensajes de Cancion1
cancion1.mostrar_datos()
cancion1.reproducir()

# Mensajes de Cancion2
cancion2.mostrar_datos()
cancion2.reproducir()

print("Programa creado por Daniel Gallegos NL = 20")
```
